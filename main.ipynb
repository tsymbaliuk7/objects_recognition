### Preparing data and creating model:


```python
import pandas as pd
from image_dataset import ImageDataset
from model import Model

folder = 'data/labels.csv'
test_folder = 'data/test_labels.csv'
data = pd.read_csv(folder, header=None)
train_dataset = ImageDataset(folder, 'data/')
test_dataset = ImageDataset(test_folder, 'data/')
model = Model(train_dataset, [{'horizontal double': 0}, {'horizontal one': 1}, {'vertical double': 2},
                                  {'vertical one': 3}], 10000, 0.001, 0.9, 0.0000001)
```

### Training model:

```python
model.train()
```

### Recognition:

```python
model.recognize(test_dataset=test_dataset)
```
